### Search for Null Values

```python
%%sql

SELECT * FROM company WHERE coalesce('CaseOrder', 'Customer_id', 'Children', 'Marital', 'Gender', 'Churn',
                                     'Contract','InternetService',
                                     'Phone', 'Multiple', 'OnlineSecurity', 'OnlineBackup',
                                     'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
                                     'PaperlessBilling', 'PaymentMethod', 'Tenure', 'MonthlyCharge') IS NULL
```

```python
%%sql

SELECT * FROM competitor WHERE coalesce('customerID', 'gender', 'Partner', 'Dependents',
                                        'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
                                        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
                                        'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
                                        'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'Churn') IS NULL
```

### Search for Duplicate Entries

```python
%%sql

SELECT "Customer_id", COUNT(*)
FROM company
GROUP BY "Customer_id"
HAVING COUNT(*) > 1
```

```python
%%sql

SELECT "customerID", COUNT(*)
FROM competitor
GROUP BY "customerID"
HAVING COUNT(*) > 1
```

#### Distinct Genders

#### Distinct Relationship Status

```python
%%sql
SELECT DISTINCT "Marital"
FROM company;
```

```python
%%sql
ALTER TABLE company
ADD COLUMN partner text;

UPDATE company
SET partner = 'No'
WHERE "Marital" IN ('Widowed', 'Never Married', 'Divorced', 'Separated');

UPDATE company
SET partner = 'Yes'
WHERE "Marital" = 'Married';
```

```python
%%sql
SELECT DISTINCT partner
FROM company;
```

```python
%%sql

ALTER TABLE company
DROP "Marital";
```

#### Distinct Dependents

```python
%%sql

ALTER TABLE company
ADD COLUMN dependents text;

UPDATE company
SET dependents = 'Yes'
WHERE "Children" > 0;

UPDATE company
SET dependents = 'No'
WHERE "Children" = 0;
```

```python
%%sql

ALTER TABLE company
DROP "Children";
```

```python
%%sql
SELECT DISTINCT dependents
FROM company;
```

```python
%%sql
SELECT DISTINCT "Dependents"
FROM competitor;
```
